<ol>
<li>Defined only for two 3D Vectors</li>
<li>Result is Another 3D Vector</li>
</ol>

<p>Now if we have two vectors in 3 dimension. Now those two vectors are defines the unique plane between them. When we compute the cross product of two vectors then we get the third vector that is defined as being orthogonal to the entire plane</p>
<img src="https://i.ibb.co/f9v1VT9/3.png" alt="3" border="0">
<p>The Third Vectors is Sometimes referred as normal to the plane</p>




**Importing the libraries**

```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

**Implementation in python Code**

```python
#Two Vectors 
v1 = np.array([-3, 2, 5])
v2 = np.array([4, -3, 0])

#Numpy's Cross product Function
v3a = np.cross(v1, v2)

#Manual Solution 
v3b = [ v1[1] * v2[2] - v1[2] * v2[1],
        v1[2] * v2[0] - v1[0] * v2[2],
        v1[0] * v2[1] - v1[1] * v2[0] ]

print("Manual Solution : \n {}".format(v3b))
print("Numpy solution : \n {}".format(v3a))
```

```python
fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection = '3d')

#drawing the plane defined by two vectors
xx , yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/ v3a[2]
ax.plot_surface(xx, yy, z1)

ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'r')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')
ax.plot([0, v3a[0]],[0, v3a[1]],[0, v3a[2]],'b')

ax.view_init(azim= 150, elev = 45)
plt.title("Plot Showing the Cross Product of Two Vectors")
plt.show()
```

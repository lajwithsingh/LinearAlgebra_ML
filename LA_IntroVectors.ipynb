<b>Importing the Libraries</b>

```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Algebraic Interpretation

#2D Dimensional Vector
v2 = [3, -1]

#3D Dimensional Vector 
v3 = [4, -3 , 2]

#Transpose of Vector v3
v3T = np.transpose(v3)
```

```python
#Geomteric Interpretation
#2Dimensional Vector

plt.plot([0, v2[0]], [0, v2[1]])
plt.axis('equal')
plt.plot([-4, 4],[0, 0], 'k--')
plt.plot([0, 0],[-4, 4], 'r--')
plt.grid()
plt.axis((-5, 5 , -5 , 5))
plt.show()
```

```python
#3Dimensional Interpretaion

fig = plt.figure(figsize=(8 , 8))
ax = fig.gca(projection = '3d')
ax.plot([0, v3[0]], [0, v3[1]], [0, v3[2]],'r',linewidth = 3)

ax.plot([0, 0],[0, 0],[-4, 4], 'k--')
ax.plot([0, 0],[-4, 4],[0, 0], 'k--')
ax.plot([-4, 4],[0, 0], [0, 0], 'k--')

plt.show()
```

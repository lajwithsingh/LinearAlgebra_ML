```python
import numpy as np
```

```python
my_matrix = np.mat([[3, 15], [9, 18]])
my_matrix
```

Symmetric Transpose

```python
my_matrix.T
```

```python
np.transpose(my_matrix)
```

Asymmetric Transpose

```python
np.transpose(-my_matrix)
```

Complex Matrix Transpose

```python
complexMat = np.matrix([[2, 9],[3, 1]], dtype = np.complex64)
complexMat
```

```python
np.conjugate(complexMat).T
```

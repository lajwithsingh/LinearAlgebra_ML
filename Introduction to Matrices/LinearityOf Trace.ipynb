```python
import numpy as np
```

```python
A = np.matrix([[3, -9, 18], [4, -9, 1],[1, 0, -19]])
A
```

```python
B = np.matrix([[6, -20, 10],[7, -18, 14],[11, 2, 3]])
B
```

```python
np.trace(A) + np.trace(B) == np.trace(A+B)
```

```python
l = 0.9
np.trace(l * A) == np.trace(A) * l
```

**Trace of Matrix is Linear Operation as It is Closed Under Addition & Multiplication**

Matrix Scalar Multiplication

```python
import numpy as np
```

```python
first_matrix = np.mat(np.array([[1, 2], [3, 4]]))
first_matrix
```

```python
first_matrix * 5
```

```python
5 * first_matrix
```

Code Challenge **Check for Distributive Property**

```python
scalar = 0.9
matrix1 = np.mat([[17, 18],[19, 21]])
matrix2 = np.mat([[21, 9],[2, 7]])

scalar * (matrix1 + matrix2)
```

```python
scalar * matrix1 + scalar * matrix2
```

**Matrix Scalar Multiplication is Linear Operation**

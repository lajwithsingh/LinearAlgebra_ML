<h1>Zoo Of Matrices</h1>

```python
import numpy as np
```

Square Vs Rectangular Matrix

```python
np.mat(np.random.randn(4,4))
```

```python
np.mat(np.random.randn(4,5))
```

Identity Matrix

```python
np.eye(5)
```

Zero Matrix

```python
np.mat(np.zeros((3,3)))
```

Digaonal Matrix

```python
np.mat(np.diag([10, 12, 17, 19, 38]))
```

Upper and Lower Triangular Matrix

```python
tempMat = np.random.randn(4,4)
```

Upper Triangular

```python
np.triu(tempMat)
```

Lower Triangular

```python
np.tril(tempMat)
```

Concatenated Matrix

```python
A = np.random.randn(4,5)
A
```

```python
B = np.random.randn(4,2)
B
```

```python
np.concatenate((A,B), axis = 1)
```

Diagonal And Trace

```python
import numpy as np
```

```python
matrix_1 = np.matrix([[2, 19],[17, -25]])
matrix_1
```

```python
np.diag(matrix_1)
```

```python
np.trace(matrix_1)
```

```python
matrix_2 = np.matrix([[3, 5, 7],[8, -9, 2]])
matrix_2
```

```python
np.diag(matrix_2)
```

```python
np.trace(matrix_2)
```

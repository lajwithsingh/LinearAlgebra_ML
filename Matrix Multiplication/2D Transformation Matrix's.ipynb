<h1>2D Transformation Matrix's</h1>

```python
#Importing the Libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
#2D input array
v = np.array([ 2, 3])

#2X2 Transformation Matrix
A = np.array([[1, -1], [2, 1]])

#The Matrix A times the vector v Av
w = A@np.matrix.transpose(v)

plt.plot([0, v[0]],[0, v[1]], label = 'v')
plt.plot([0, w[0]],[0, w[1]], label = 'Av')
plt.grid()
plt.legend()
plt.axis((-4, 4, -2, 6))
plt.title("Rotation + Stretching")
plt.show()
```

```python
#2D input array
v = np.array([ 2, 3])

#2X2 Pure Rotation Matrix
theta = -20
A = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

#The Matrix A times the vector v Av
w = A@np.matrix.transpose(v)

plt.plot([0, v[0]],[0, v[1]], label = 'v')
plt.plot([0, w[0]],[0, w[1]], label = 'Av')
plt.grid()
plt.legend()
plt.axis((-4, 4, -4, 4))
plt.title("Rotation + Stretching")
plt.show()
```

```python
#importing the libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
length_vector = []

#2D vector
v = np.array([2, -9])


for div in range(180,0,-30):
    #theta value
    theta = np.pi/div

    #Pure Rotation Matrix
    pure_rot = np.matrix([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
    
    #vector times rotation matrix
    w = pure_rot@np.matrix.transpose(v)
    
    length_vector.append(np.linalg.norm(w))
    
plt.plot([x for x in range(180, 0, -30)],length_vector,'o-')
plt.grid()
plt.title("Matrix Transformation with Pure Matrix")
plt.show()
```

```python
length_vector = []

#2D vector
v = np.array([2, -9])


for div in range(180,0,-30):
    #theta value
    theta = np.pi/div

    #Impure Rotation Matrix
    impure_rot = np.matrix([[np.cos(theta) * 2, -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
    
    #vector times rotation matrix
    w = impure_rot@np.matrix.transpose(v)
    
    length_vector.append(np.linalg.norm(w))
    
plt.plot([x for x in range(180, 0, -30)],length_vector,'o-')
plt.grid()
plt.title("Matrix Transformation with Impure Matrix")
plt.show()
```
